```python
import cv2
import sys
from google.colab.patches import cv2_imshow
```

```python
img_path="/content/abba.png"
module_path="/content/haarcascade_frontalface_default.xml"
face=cv2.CascadeClassifier(module_path)
```

```python
img_open=cv2.imread(img_path)
cv2_imshow(img_open)
```

```python
img_gray=cv2.cvtColor(img_open,cv2.COLOR_BGR2GRAY)
faces=face.detectMultiScale(img_gray,minSize=(30,30))
print(len(faces))
```

```python
for (x,y,w,h) in faces :
  cv2.rectangle(img_open,(x,y),(x+w,y+h),(0,0,255),2)

cv2_imshow(img_open)
```
